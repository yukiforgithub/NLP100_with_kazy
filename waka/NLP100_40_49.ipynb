```python
idx_list=[]
dst_list=[]
idx_list_list=[]
dst_list_list=[]
src_dict_list=[]
for i in text_list_ini:
    if "* " in i[0]:
        cabo_res=i[0].split(" ")
        idx=cabo_res[1]
        dst=cabo_res[2][:-1]
        idx_list.append(idx)
        dst_list.append(dst)
    if "EOS" == i[0]:
        idx_list_list.append(idx_list)
        dst_list_list.append(dst_list)
        src_dict={}
        for idx_list_idx in range(len(idx_list_list)):
            dst_list=dst_list_list[idx_list_idx]
            for idx in idx_list_list[idx_list_idx]:
                where=np.where(idx==np.array(dst_list))
                if len(where[0])==0:
                    continue
                src_dict[idx]=list(where[0])
        src_dict_list.append(src_dict)
        idx_list=[]
        dst_list=[]
```
